```python
import pandas as pd

from sklearn import datasets

from evidently.report import Report
from evidently.metric_preset import DataDriftPreset

iris_data = datasets.load_iris(as_frame='auto')
iris_frame = iris_data.frame
```

```python
iris_frame.head()
```

```python
data_drift_report = Report(metrics=[
    DataDriftPreset(),
])

data_drift_report.run(current_data=iris_frame.iloc[:90], reference_data=iris_frame.iloc[90:], column_mapping=None)
data_drift_report
```

```python
data_drift_report.save_html("drift.html")
```

```python
from evidently.metric_preset import DataQualityPreset, RegressionPreset
from evidently import ColumnMapping
```

```python
set(iris_frame['target'])
```

```python
data_quality_report = Report(metrics=[
    DataQualityPreset()
])

column_mapping = ColumnMapping()
data_quality_report.run(current_data=iris_frame, reference_data=None, column_mapping=None)
data_quality_report
```

```python
data_quality_report.save_html('quality.html')
```
